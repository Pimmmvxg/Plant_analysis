```python
import matplotlib
from plantcv import plantcv as pcv
from plantcv.utils import tabulate_bayes_classes
from plantcv.parallel import WorkflowInputs
import numpy as np
from plantcv.plantcv.analyze import color as analyze_color
```

```python
args = WorkflowInputs(
    images=[r"C:\Plant_analysis\notebooks\datasets\top_view\multi2.jpg"],
    names="top_view",
    result="top_view.json",
    debug="plot"
)
pcv.params.debug = args.debug
pcv.params.dpi = 100
pcv.params.text_size = 10
pcv.params.text_thickness = 20
```

```python
color_img, path, filename = pcv.readimage(filename=args.top_view)
```

```python
colorspaces = pcv.visualize.colorspaces(rgb_img=color_img, original_img=False)
```

```python
h = pcv.rgb2gray_lab(rgb_img=color_img, channel='a')
```

```python
thresh = pcv.threshold.gaussian(gray_img=h, ksize=9000, offset=10,
                              object_type='light')
```

```python
thresh = pcv.fill(bin_img=thresh, size=400)
```

```python
rois1 = pcv.roi.multi(img=color_img, coord=(100,200), radius=200,
                      spacing=(500,400), nrows=3, ncols=2)
```

```python
lbl_mask, n_lbls = pcv.create_labels(mask=thresh, rois=rois1)
```

```python
analyze_color(rgb_img=color_img, labeled_mask=lbl_mask, n_labels=n_lbls, colorspaces='all')
```

```python
pcv.outputs.save_results(filename="output.json") 

results_dict = pcv.outputs.observations
print(results_dict)
```
